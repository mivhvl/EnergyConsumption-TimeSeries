### 2.4 Stationarity Check

The next step is to check for stationarity, which is important for time series models like ARIMA. The Augmented Dickey-Fuller (ADF) test helps determine if the time series is stationary by testing the null hypothesis of a unit root. If the p-value is less than or equal to 0.05, we reject the null hypothesis and conclude the series is stationary. If the p-value is greater than 0.05, the series is non-stationary and may need transformation before modeling.

```python
# Function to perform ADF test
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value}")
    if result[1] <= 0.05:
        print("Reject the null hypothesis (Stationary)")
    else:
        print("Fail to reject the null hypothesis (Non-stationary)")

adf_test(monthly_data['AEP_MW'])

def check_stationarity(timeseries):
    # Perform the Dickey-Fuller test
    result = adfuller(timeseries, autolag='AIC')
    p_value = result[1]
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {p_value}')
    print('Stationary' if p_value < 0.05 else 'Non-Stationary')

check_stationarity(monthly_data['AEP_MW'])
```

The p-value of 0.43 means we fail to reject the null hypothesis, indicating the series is non-stationary. The ADF statistic is also higher than the critical values, confirming the data is not stationary.

### 2.5 ACF and PACF Diagrams

We use the Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) to identify the relationships within a time series. ACF measures the overall correlation between a time series and its lagged values, while PACF isolates the correlation between the series and a specific lag. These plots are essential for determining the order of ARIMA models.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import numpy as np
import matplotlib.pyplot as plt

# Calculate sample size dynamically
n = len(monthly_data['AEP_MW'].dropna())

# Confidence interval (not directly used here since the plot includes it)
conf_interval = 1.96 / np.sqrt(n)

plt.figure(figsize=(12, 6))

# ACF plot
plt.subplot(121)
plot_acf(monthly_data['AEP_MW'].dropna(), lags=50, ax=plt.gca())
plt.title('ACF of the Time Series')
plt.xlabel('Lags', fontsize=12)
plt.ylabel('Autocorrelation', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# PACF plot
plt.subplot(122)
plot_pacf(monthly_data['AEP_MW'].dropna(), lags=50, ax=plt.gca(), method='ywm')
plt.title('PACF of the Time Series')
plt.xlabel('Lags', fontsize=12)
plt.ylabel('Partial Autocorrelation', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### 4.1. SARIMA Implementation

Because of its strong seasonality, as we saw in the Seasonal Decomposition, the decision to use a SARIMA model was made. To build this model, we need to understand not only the non-seasonal component, but also the seasonal components.

#### Non-Seasonal Components:

- **d (Degree of Diferencing)**: This determines the number of times the series needs to be differenced to make it stationary; As proved by the ADF Test, the Time Series is non-stationary. We used **d = 1** because this is a reasonable starting point for removing trends in many real-world datasets

- **p (Autoregressive Order)**: This indicates the number of significant lags in the PACF plot before the curve diminishes; From the PACF plot, there is a significant spike at lag 1, suggesting that **p = 1**

- **q (Moving Average Order)**: Finally, this is the number of significant lags in the ACF plot before the curve diminishes; From the ACF plot, there is a significant spike at lag 1, suggesting that **q = 1**

#### Seasonal Components:

**D (Seasonal Diferencing)**: From the seasonal decomposition, the time series exhibits strong seasonality with a clear yearly cycle. Seasonal differencing **(D = 1)** is required to address this periodicity

**s (Seasonal Periodicity)**: The seasonal period is determined by the frequency of the dataset; Since this is monthly data, the seasonal period is **s = 12**

**P (Seasonal Autoregressive Order)**: Significant lags in the PACF plot at multiples of the seasonal period **(s = 12)**; From the PACF plot, lag 12 shows significance, suggesting **P = 1**

**Q (Seasonal Moving Average Order)**: Significant lags in the ACF plot at multiples of the seasonal period **(s = 12)**; Like the PACF plot, lag 12 shows significance, suggesting **Q = 1**

```python
# Ensure no SettingWithCopyWarning
df_train = df_train.copy()  # Optional: if df_train is a slice of a larger DataFrame
df_train['AEP_MW'] = df_train['AEP_MW'].fillna(method='ffill')

# Fit SARIMA model with differencing
model = SARIMAX(df_train['AEP_MW'], 
                order=(1, 1, 1),  # p, d, q for AR, differencing, MA
                seasonal_order=(1, 1, 1, 12),  # P, D, Q, m for seasonal components (12 for monthly data)
                enforce_stationarity=False, 
                enforce_invertibility=False)

# Fit model
model_fit = model.fit(disp=False)

# Print summary
print(model_fit.summary())
# Number of steps to forecast (aligned with the test set)
forecast_steps = len(df_test)

# Forecast future values for the test set
forecast = model_fit.get_forecast(steps=forecast_steps)

# Get the predicted values and confidence intervals
forecast_mean = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Plot the training data, test data, and forecast
plt.figure(figsize=(12, 6))
plt.plot(df_train['AEP_MW'], label='Training Data', color='orange', linewidth=3)
plt.plot(df_test['AEP_MW'], label='Test Data', color='grey', linewidth=3)
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='blue', linestyle='--', linewidth=2)
plt.fill_between(forecast_mean.index, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], color='blue', alpha=0.2)

plt.title('SARIMA Forecast', fontsize=15)
plt.xlabel('Date', fontsize=12)
plt.ylabel('AEP_MW', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### 4.2.3. Ljung-Box Test

The Ljung-Box test assesses whether the residuals of a time series model are independently distributed (i.e., no autocorrelation). A high p-value (> 0.05) indicates that the residuals resemble white noise, suggesting the model has captured the autocorrelation effectively.

```python
# Ljung-Box Test
residuals = residuals.dropna()
lb_test = acorr_ljungbox(residuals, lags=[10], model_df=4, return_df=True)
print(lb_test)
```

The Null Hypothesis (\(H_0\)) means "Residuals are not autocorrelated" and we had p-value = 0.622. 
Since p > 0.05, we fail to reject \(H_0\), indicating no significant autocorrelation in the residuals.

The SARIMA model performs well in capturing the temporal dependencies.

### 4.2.3. Cross-Validation

- **TimeSeriesSplit** is an appropriate choice for time-series data since it respects the temporal order by splitting the data into sequential training and test sets.
- The use of RMSE as the error metric provides an interpretable measure of the model's prediction accuracy.

```python
warnings.filterwarnings("ignore")

# Define the number of splits for cross-validation
tscv = TimeSeriesSplit(n_splits=5)

# Initialize a list to store errors
errors = []

# Perform cross-validation
for train_index, test_index in tscv.split(df_train):
    train, test = df_train.iloc[train_index], df_train.iloc[test_index]
    
    # Fit the SARIMA model on the training data
    model = SARIMAX(train['AEP_MW'], 
                    order=(1, 1, 1), 
                    seasonal_order=(1, 1, 1, 12), 
                    enforce_stationarity=False, 
                    enforce_invertibility=False)
    model_fit = model.fit(disp=False)
    
    # Forecast on the test set
    forecast = model_fit.forecast(steps=len(test))
    
    # Calculate error (e.g., RMSE)
    error = np.sqrt(mean_squared_error(test['AEP_MW'], forecast))
    errors.append(error)

# Calculate the average error across splits
mean_error = np.mean(errors)
print(f'Mean RMSE from Cross-Validation: {mean_error}')
```

- **Mean RMSE**: \(1080.28\) MW indicates the average error between the predicted and actual test values during cross-validation.

- **Interpretation**: While this error magnitude seems reasonable, it should be compared to the mean or variance of the original data to gauge relative accuracy. If the error is a small fraction of the typical values in the time series, the model can be considered effective.
